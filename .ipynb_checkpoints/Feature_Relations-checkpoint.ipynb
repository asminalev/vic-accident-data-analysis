```python
def compare_with_target(df: pd.DataFrame, col: str) -> None: 
    
    '''Graph column 'col' of dataframe 'df' grouped by TARGET column.
     Generate boxplot and histogram plot for int64 or float64 column and bar count plot
     (countplot) for columns of other types except the TARGET column.
     Args:
         df (pd.DataFrame): dataframe with 'col'`s to be plotted.
         col(str): string with column name.
    '''
    
    
    
    if df[col].dtype in ['int64', 'float64']:
        _, axs = plt.subplots(1, 2, figsize=(15, 10))
        sns.boxplot(
            data=df.melt(id_vars='severity', value_vars=col),
            y='variable', x='value', hue='severity', ax=axs[0], showmeans=True, showfliers = False
            )
        axs[0].set_title('Boxplot')
        sns.histplot(data=df, x=col, ax=axs[1], kde=True, hue='severity')
        xlim_min = 0
        xlim_max = df[col].std()*1.75
        plt.xlim(xlim_min,xlim_max)


        axs[1].set_title('Histogram')
    elif col != 'severity':
        sns.countplot(data=df, x=col, hue='severity')
```

```python
compare_with_target(df_selected_columns, 'day_week_description')
```
